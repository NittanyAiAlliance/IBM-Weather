### TC Count and Indexs Visualization

```python
# Get IBTrACS
import xarray as xr
import netCDF4 as nc
ibtracs = pd.read_csv("../Data/TC Count/IBTrACS/IBTrACS(1980-P, 100W-20W&10N-40N).csv")
```

```python
# Read Nino3.4, QBO, SOI
nino = pd.read_csv("../Data/Indexs/Nino3.4.txt", sep=" ",names = [" ","Year in Decimal","Nino index","1 "]).dropna(axis = 1, how = 'all').dropna(axis = 0, how = 'any')
soi = pd.read_csv("../Data/Indexs/SOI.txt", sep=" ",names = [" ","Year in Decimal","SOI index","1 "]).dropna(axis = 1, how = 'all').dropna(axis = 0, how = 'any')
qbo = pd.read_csv("../Data/Indexs/QBO.txt", sep=" ",names = [" ","Year in Decimal","QBO index","1 "]).dropna(axis = 1, how = 'all').dropna(axis = 0, how = 'any')
```

```python
def convert_partial_year(number):
    year = int(number)
    d = timedelta(days = ((number - year) * 365)+2)
    day_one = datetime(year, 1, 1)
    date = d + day_one
    return date
nino["Year in dt"] = nino["Year in Decimal"].apply(convert_partial_year)
nino["Year"] = nino["Year in dt"].dt.year
nino["Month"] = nino["Year in dt"].dt.month
nino = nino.loc[nino['Year'] >= 1950]
soi["Year in dt"] = soi["Year in Decimal"].apply(convert_partial_year)
soi["Year"] = soi["Year in dt"].dt.year
soi["Month"] = soi["Year in dt"].dt.month
soi = soi.loc[soi['Year'] >= 1950]
qbo["Year in dt"] = qbo["Year in Decimal"].apply(convert_partial_year)
qbo["Year"] = qbo["Year in dt"].dt.year
qbo["Month"] = qbo["Year in dt"].dt.month
qbo = qbo.loc[qbo['Year'] >= 1950]
```

```python
# Calculate Norms
nino_norm = (np.array(nino["Nino index"]) - np.min(np.array(nino["Nino index"]))) / (np.max(np.array(nino["Nino index"])) - np.min(np.array(nino["Nino index"])))
soi_norm = (np.array(soi["SOI index"]) - np.min(np.array(soi["SOI index"]))) / (np.max(np.array(soi["SOI index"])) - np.min(np.array(soi["SOI index"])))
qbo_norm = (np.array(qbo["QBO index"]) - np.min(np.array(qbo["QBO index"]))) / (np.max(np.array(qbo["QBO index"])) - np.min(np.array(qbo["QBO index"])))
def moving_average(x, w):
    return np.convolve(x, np.ones(w), 'same') / w
nino_rollingMean = moving_average(nino_norm,12)
soi_rollingMean = moving_average(soi_norm,12)
qbo_rollingMean = moving_average(qbo_norm,12)
```

```python
fig, axs = plt.subplots(5,1)
fig.set_size_inches(10,20)
axs[0].plot(nino["Year in dt"],nino["Nino index"],c = 'y')
axs[0].set_title('Nino Index')
axs[1].plot(soi["Year in dt"],soi["SOI index"],c = 'b')
axs[1].set_title('SOI Index')
axs[2].plot(qbo["Year in dt"],qbo["QBO index"],c = 'orange')
axs[2].set_title('QBO Index')
axs[3].plot(nino["Year in dt"],nino_norm,c = 'y')
axs[3].plot(soi["Year in dt"],soi_norm,c = 'b')
axs[3].plot(qbo["Year in dt"],qbo_norm,c = 'orange')
axs[3].set_title('Norms')
axs[4].plot(nino["Year in dt"],nino_rollingMean,c = 'y',label = 'nino3.4')
axs[4].plot(soi["Year in dt"],soi_rollingMean,c = 'b',label = 'soi')
axs[4].legend()
axs[4].set_title('Rolling Mean')
```

```python
### Histogram of SOI, Nino, QBO
```

```python
#startingYear = np.max([np.min(soi['Year']),np.min(nino['Year']),np.min(qbo['Year'])])
#endingYear = np.min([np.max(soi['Year']),np.max(nino['Year']),np.max(qbo['Year'])])
startingYear = 1980
endingYear = 2022
soi_yearAvg = np.zeros(endingYear - startingYear + 1)
nino_yearAvg = np.zeros(endingYear - startingYear + 1)
qbo_yearAvg = np.zeros(endingYear - startingYear + 1)
for i in range(startingYear,endingYear+1):
    soi_temp = soi.loc[(soi["Year"] == i) & (soi['Month'] <= 6) & (soi['Month'] >= 3)]["SOI index"]
    nino_temp = nino.loc[(nino["Year"] == i) & (nino['Month'] <= 6) & (nino['Month'] >= 3)]["Nino index"]
    qbo_temp = qbo.loc[(qbo["Year"] == i) & (qbo['Month'] <= 6) & (qbo['Month'] >= 3)]["QBO index"]
    soi_yearAvg[i-endingYear] = np.mean(soi_temp)
    nino_yearAvg[i-endingYear] = np.mean(nino_temp)
    qbo_yearAvg[i-endingYear] = np.mean(qbo_temp)
nino_yearNorm = (nino_yearAvg - np.min(nino_yearAvg)) / (np.max(nino_yearAvg) - np.min(nino_yearAvg))
soi_yearNorm = (soi_yearAvg - np.min(soi_yearAvg)) / (np.max(soi_yearAvg) - np.min(soi_yearAvg))
qbo_yearNorm = (qbo_yearAvg - np.min(qbo_yearAvg)) / (np.max(qbo_yearAvg) - np.min(qbo_yearAvg))
TCCount_temp = ibtracs.loc[ibtracs['SEASON'] >= startingYear]['COUNT']
TCCount_yearNorm = (TCCount_temp - np.min(TCCount_temp)) / (np.max(TCCount_temp) - np.min(TCCount_temp))
```

```python
plt.figure(figsize=(20,5))
plt.plot(np.arange(startingYear, endingYear+1), nino_yearNorm, '--',lineWidth = 3, label = "NINO 3.4")
plt.plot(np.arange(startingYear, endingYear+1), qbo_yearNorm,'--',lineWidth = 3, label = "QBO")
plt.plot(np.arange(startingYear, endingYear+1), soi_yearNorm,'--',lineWidth = 3, label = "SOI")
plt.plot(ibtracs.loc[ibtracs['SEASON'] >= startingYear]['SEASON'], TCCount_yearNorm, lineWidth = 5,label = "TC Count")
plt.xticks(np.arange(startingYear,endingYear+1,2))
plt.legend()
plt.show()
```

```python
dataframe = pd.DataFrame({'NINO3.4': nino_yearNorm[:-2],'QBO':qbo_yearNorm[:-2],'SOI':soi_yearNorm[:-2],'TC Count':TCCount_yearNorm})
```

### Atomspheric Features Visualization

```python
finalFeatDF = pd.read_csv("../Data/MERRA2/Features.csv")
```

```python
finalFeatDF.describe()
```

```python
finalFeatDF = finalFeatDF.drop(['Unnamed: 0'],axis = 1)
```

```python
fig, axs = plt.subplots(2,2)
fig.set_size_inches(20,10)
axs[0,0].hist(finalFeatDF['OMEGA500'],bins = 20,range = [-0.2,0.2],)
axs[0,0].set_title('OMEGA at 500hPa')
axs[0,1].hist(finalFeatDF['SLP'],bins = 20,range = [98000,103000])
axs[0,1].set_title('Sea Level Pressure')
axs[1,0].hist(finalFeatDF['U500'],bins = 20,)
axs[1,0].set_title('Eastward Wind at 500 hPa')
axs[1,1].hist(finalFeatDF['SSSMASS'],bins = 20,)
axs[1,1].set_title('Sea Surface Salt Mass Concentration')
```

```python
### coordinates for 3d plot
lon = finalFeatDF['lon'].unique()
lat = finalFeatDF['lat'].unique()
X,Y = np.meshgrid(lon,lat)

fig = plt.figure()
fig.set_size_inches(20,20)

### Get z values
features = ['OMEGA500','SLP','U500','SSSMASS']
year = 2020
month = 5    
for k in range(len(features)):
    Z = np.zeros((X.shape[0],X.shape[1]))
    feature = features[k]
    for i in range(Z.shape[0]):
        for j in range(Z.shape[1]):
            target_cord = [X[i,j],Y[i,j]]
            Z[i,j] = finalFeatDF.loc[(finalFeatDF['year'] == year) &
                                     (finalFeatDF['month'] == month) &
                                     (finalFeatDF['lat'] == target_cord[1])&
                                     (finalFeatDF['lon'] == target_cord[0])][features[k]]
    ### Plot
    ax = fig.add_subplot(2,2,k+1)
    cf = ax.contourf(X, Y, Z,30,cmap = 'cool')
    ax.set_xlabel('lon')
    ax.set_ylabel('lat')
    plt.colorbar(cf)
```

```python
from pylab import *
colo = [X + Y + Z]
  
# creating figures
fig = plt.figure(figsize=(10, 10))
ax = fig.add_subplot(111, projection='3d')
  
# setting color bar
color_map = cm.ScalarMappable(cmap=cm.Greens_r)
color_map.set_array(colo)
  
# creating the heatmap
img = ax.scatter(X, Y, Z, marker='s',
                 s=200)
plt.colorbar(color_map)
  
# adding title and labels
ax.set_title("3D Heatmap")
ax.set_xlabel('X-axis')
ax.set_ylabel('Y-axis')
ax.set_zlabel('Z-axis')
  
# displaying plot
plt.show()
```
